# Wave Equation

The wave equation is a partial differential equation that describes the propagation of waves through a medium. The equation takes the form:

$$\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2}$$

where $u$ is the displacement of the wave, $c$ is the speed of the wave, $t$ is time, and $x$ is the position along the medium. The wave equation is a second-order linear partial differential equation and as such can be solved using the method of separation of variables. The general solution of the wave equation is a superposition of the right- and left-traveling waves:

$$u(x,t) = f(x-ct) + g(x+ct)$$

where $f$ and $g$ are arbitrary functions. The first term describes a wave traveling to the right with speed $c$, while the second describes a wave traveling to the left with speed $c$.

The wave equation is also known as the one-dimensional wave equation to distinguish it from the three-dimensional wave equation.

## Derivation

Since each of the time and position variables are of second order, we expect to have two sets of initial conditions for each of time and space.

$$u(x,0) = f(x)$$
$$\frac{\partial u}{\partial t}(x,0) = g(x)$$

If the string is fixed at both ends, then the boundary conditions are:

$$u(0,t) = 0 $$
$$u(L,t) = 0 $$

where $L$ is the length of the string.

Let's seek solutions of the form:

$$u(x,t) = e^{ikx + \sigma t}$$

where $k$ and $\sigma$ are constants. Substituting this into the wave equation gives:

$$\sigma^2 e^{ikx + \sigma t} = c^2 (ik)^2 e^{ikx + \sigma t} = - c^2 k^2 e^{ikx + \sigma t} $$

which simplifies to:

$$\sigma^2 = - c^2 k^2$$

or:

$$\sigma = \pm i c k$$

The general solution is a superposition of the right- and left-traveling waves:

$$u(x,t) = e^{ik(x \pm ct)}$$

where $k$ is an arbitrary constant. The '+' sign describes a wave traveling to the right with speed $c$, while the '-' describes a wave traveling to the left with speed $c$.

## D'Alembert's Solution

D'Alembert's solution is a general solution of the wave equation that is valid for any initial conditions and boundary conditions. 

We begin with the following change of variables:

$$ r = x + ct $$
$$ s = x - ct $$
$$ x = \frac{r+s}{2} $$
$$ t = \frac{r-s}{2c} $$

Then, the partial derivatives with respect to $r$ and $t$ are:

$$ \frac{\partial}{\partial r} = \frac{\partial}{\partial x} \frac{\partial x}{\partial r} + \frac{\partial}{\partial t} \frac{\partial t}{\partial r} = \frac{\partial}{\partial x} \frac{1}{2} + \frac{\partial}{\partial t} \frac{1}{2c} = \frac{1}{2} \frac{\partial}{\partial x} + \frac{1}{2c} \frac{\partial}{\partial t} $$

$$ \frac{\partial}{\partial s} = \frac{\partial}{\partial x} \frac{\partial x}{\partial s} + \frac{\partial}{\partial t} \frac{\partial t}{\partial s} = \frac{\partial}{\partial x} \frac{1}{2} - \frac{\partial}{\partial t} \frac{1}{2c} = \frac{1}{2} \frac{\partial}{\partial x} - \frac{1}{2c} \frac{\partial}{\partial t} = 0$$

Therefore,

$$-4c^2 \frac{\partial^2 u}{\partial r \partial s} = -4c^2 \frac{\partial}{\partial r} \frac{\partial}{\partial s} u = -4c^2 \left( \frac{1}{2} \frac{\partial}{\partial x} + \frac{1}{2c} \frac{\partial}{\partial t} \right) \left( \frac{1}{2} \frac{\partial}{\partial x} - \frac{1}{2c} \frac{\partial}{\partial t} \right) u = -c^2 \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial t^2}$$

Thus, we have that:

$$\frac{\partial^2 u}{\partial r \partial s} = 0$$

Integrating with respect to $r$ gives:

$$\frac{\partial u}{\partial s} = \bar{\phi}_1(s)$$

where $\bar{\phi}_1(s)$ is an arbitrary function. Integrating again with respect to $s$ gives:

$$u(r,s) = \int \bar{\phi}_1(s) ds + \phi_2(r) = \phi_1(s) + \phi_2(r)$$

## Initial Conditions

Suppose that we have the following initial conditions:

$$u(x,0) = u_0(x)$$
$$\frac{\partial u}{\partial t}(x,0) = v_0(x)$$

Then, we have that:

$$u(x,t) = F(x-ct) + G(x+ct) = F(x) + G(x) = u_0(x)$$

$$\frac{\partial u}{\partial t}(x,t) = -cF'(x-ct) + cG'(x+ct) = -cF'(x) + cG'(x) = v_0(x)$$

Therefore,

$$-cF(x) + cG(x) = \int_0^x v_0(\xi) d\xi + A$$

Solving these two equations in two unknowns for $F$ and $G$, and then substituting into the general solution gives:

$$u(x,t) = \frac{1}{2} \left[ u_0(x+ct) + u_0(x-ct) \right] + \frac{1}{2c} \int_{x-ct}^{x+ct} v_0(\xi) d\xi$$

## Numerical Solution

The wave equation can be solved numerically using the finite difference method. The second-order partial derivatives can be approximated using the central difference formula:

$$\frac{\partial^2 u}{\partial t^2} \approx \frac{u(x,t+\Delta t) - 2u(x,t) + u(x,t-\Delta t)}{\Delta t^2}$$

$$\frac{\partial^2 u}{\partial x^2} \approx \frac{u(x+\Delta x,t) - 2u(x,t) + u(x-\Delta x,t)}{\Delta x^2}$$

Substituting these into the wave equation gives:

$$\frac{u(x,t+\Delta t) - 2u(x,t) + u(x,t-\Delta t)}{\Delta t^2} = c^2 \frac{u(x+\Delta x,t) - 2u(x,t) + u(x-\Delta x,t)}{\Delta x^2}$$

Solving for $u(x,t+\Delta t)$ gives:

$$u(x,t+\Delta t) = 2u(x,t) - u(x,t-\Delta t) + \left( \frac{c\Delta t}{\Delta x} \right)^2 \left[ u(x+\Delta x,t) - 2u(x,t) + u(x-\Delta x,t) \right]$$


# Example:  Vibrating String

Consider a string of length $L$ fixed at both ends. The initial displacement of the string is given by:

$$u(x,0) = \sin \left( \frac{\pi x}{L} \right)$$

and the initial velocity of the string is given by:

$$\frac{\partial u}{\partial t}(x,0) = 0$$

The boundary conditions are:

$$u(0,t) = 0 $$
$$u(L,t) = 0 $$

The solution to the wave equation is:

$$u(x,t) = \frac{1}{2} \left[ \sin \left( \frac{\pi (x+ct)}{L} \right) + \sin \left( \frac{\pi (x-ct)}{L} \right) \right]$$

```python
# Numerical solution of the wave equation

import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
L = 1.0
c = 1.0
T = 1.0
Nx = 100
Nt = 1000

# Set the grid spacing
dx = L/Nx
dt = T/Nt

# Stability condition

if dt > dx/c:
    print('Stability condition not satisfied!')
    print('dt = ', dt)
    print('dx/c = ', dx/c)
    print('alpha^2 = ', (c*dt/dx)**2)
    print('Try again with smaller dt or larger dx')
    exit()
else:
    print('Stability condition satisfied!')
    print('dt = ', dt)
    print('dx/c = ', dx/c)
    print('alpha^2 = ', (c*dt/dx)**2)
    print('Proceeding with the calculation')
    
# Define the initial condition function
def u0(x):
    global L
    return np.sin(3*np.pi*x/L)

# Set the grid
x = np.linspace(0, L, Nx+1)
t = np.linspace(0, T, Nt+1)

# Set the initial conditions and boundary conditions
u = np.zeros((Nx+1, Nt+1))
v = np.zeros((Nx+1, Nt+1))
u[:,0] = u0(x)
v[:,0] = 0

u[0,:] = 0
u[Nx,:] = 0

# Calculate the solution for the first time step
for i in range(1, Nx):
    u[i,1] = dt*v[i,0] + u[i,0]  + (c*dt/dx)**2 * (u[i+1,0] - 2*u[i,0] + u[i-1,0])

# Solve the wave equation using the finite difference method for the second time step and beyond

for n in range(1, Nt):
    for i in range(1, Nx):
        u[i,n+1] = 2*u[i,n] - u[i,n-1] + (c*dt/dx)**2 * (u[i+1,n] - 2*u[i,n] + u[i-1,n])
        
# Calculate the exact solution
u_exact = np.zeros((Nx+1, Nt+1))
for n in range(0, Nt+1):
    for i in range(0, Nx+1):
        u_exact[i,n] = 0.5 * (u0(x[i]+c*t[n]) + u0(x[i]-c*t[n]))

fig, ax = plt.subplots(2,2, figsize=(10,10))

ax[0][0].plot(x, u_exact[:,Nt//4], label='t=T/4 Exact')
ax[0][0].plot(x, u[:,Nt//4], '--',label='t=T/4')

ax[0][1].plot(x, u_exact[:,Nt//2], label='t=T/2 Exact')
ax[0][1].plot(x, u[:,Nt//2], '--',label='t=T/2')

ax[1][0].plot(x, u_exact[:,3*Nt//4], label='t=3T/4 Exact')
ax[1][0].plot(x, u[:,3*Nt//4], '--',label='t=3T/4')

ax[1][1].plot(x, u_exact[:,Nt], label='t=T Exact')
ax[1][1].plot(x, u[:,Nt], '--',label='t=T')

ax[0][0].set_xlabel('x')
ax[0][0].set_ylabel('u')
ax[0][0].set_ylim([-1.1,1.1])

ax[0][1].set_xlabel('x')
ax[0][1].set_ylabel('u')
ax[0][1].set_ylim([-1.1,1.1])

ax[1][0].set_xlabel('x')
ax[1][0].set_ylabel('u')
ax[1][0].set_ylim([-1.1,1.1])

ax[1][1].set_xlabel('x')
ax[1][1].set_ylabel('u')
ax[1][1].set_ylim([-1.1,1.1])

ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```

```python
# Create a 3D plot of the solution

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
T, X = np.meshgrid(t, x)
ax.plot_surface(T, X, u_exact, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('u')
plt.show()
```
