# Complex solutions to algebraic equations

The fundamental theorem of algebra states that every polynomial equation has at least one solution.  For example, the equation $x^2 + 1 = 0$ has no real solutions, but it has two complex solutions, $x = i$ and $x = -i$.  The complex numbers are an extension of the real numbers that includes a solution to this equation, and many others. 

As an examples, consider the following problem:

> Find the roots of the equation $x^6 = (x-1)^6$.

## Analytic solution

The analytic solution to this problem is interesting, and involves a number of steps.  First, we write the equation as:

$$x^6 - (x-1)^6 = 0$$

Next, we can use the formula for the difference of two squares to factor the left hand side:

$$\left(x^3 - (x-1)^3\right)\left(x^3 + (x-1)^3\right) = 0$$

Next, we can use the formula for the difference of two cubes to factor the first term:

$$\left(x - (x-1)\right)\left(x^2 + x(x-1) + (x-1)^2\right)\left(x^3 + (x-1)^3\right) = 0$$

Next, we can use the formula for the sum of two cubes to factor the second term:

$$\left(x - (x-1)\right)\left(x^2 + x(x-1) + 
(x-1)^2\right)\left(x + (x-1)\right)\left(x^2 - x(x-1) + 
(x-1)^2\right) = 0$$

Simplifying, we get:

$$\left(x^2 + x^2 - x + x^2 - 2x + 1\right)\left(2x -1\right)\left(x^2 - x^2 + x + x^2 - 2x + 1\right) = 0$$
$$\left(3x^2 - 3x + 1\right)\left(2x -1\right)\left(x^2 - x + 1\right) = 0$$

This now has three factors, each of which can be set to zero to find the roots:

$$3x^2 - 3x + 1 = 0$$
$$2x - 1 = 0$$
$$x^2 - x + 1 = 0$$

The first equation can be solved using the quadratic formula:

$$x = \frac{3 \pm \sqrt{9 - 12}}{6} = \frac{3 \pm \sqrt{-3}}{6} = \frac{3 \pm i\sqrt{3}}{6}$$

The second equation can be solved by adding 1/2 to both sides:

$$x = \frac{1}{2}$$

The third equation can be solved using the quadratic formula:

$$x = \frac{1 \pm \sqrt{1 - 4}}{2} = \frac{1 \pm \sqrt{-3}}{2} = \frac{1 \pm i\sqrt{3}}{2}$$


```python
# Plot the roots

import numpy as np
import matplotlib.pyplot as plt

x1 = 1/2
x2 = (3 + 1j*np.sqrt(3))/6
x3 = (3 - 1j*np.sqrt(3))/6
x4 = (1 + 1j*np.sqrt(3))/2
x5 = (1 - 1j*np.sqrt(3))/2

theta1 = np.arctan(x1.imag/x1.real) * 180/np.pi
theta2 = np.arctan(x2.imag/x2.real) * 180/np.pi
theta3 = np.arctan(x3.imag/x3.real) * 180/np.pi
theta4 = np.arctan(x4.imag/x4.real) * 180/np.pi
theta5 = np.arctan(x5.imag/x5.real) * 180/np.pi

y1 = x1**6/abs(x1**6)
y2 = x2**6/abs(x2**6)
y3 = x3**6/abs(x3**6)
y4 = x4**6/abs(x4**6)
y5 = x5**6/abs(x5**6)

z1 = (x1-1)**6/abs((x1-1)**6)
z2 = (x2-1)**6/abs((x2-1)**6)
z3 = (x3-1)**6/abs((x3-1)**6)
z4 = (x4-1)**6/abs((x4-1)**6)
z5 = (x5-1)**6/abs((x5-1)**6)


print('x1 = {:.3f} + {:.3f}j, theta = {:.3f}'.format(x1.real, x1.imag, theta1))
print('x2 = {:.3f} + {:.3f}j, theta = {:.3f}'.format(x2.real, x2.imag, theta2))
print('x3 = {:.3f} + {:.3f}j, theta = {:.3f}'.format(x3.real, x3.imag, theta3))
print('x4 = {:.3f} + {:.3f}j, theta = {:.3f}'.format(x4.real, x4.imag, theta4))
print('x5 = {:.3f} + {:.3f}j, theta = {:.3f}'.format(x5.real, x5.imag, theta5))

print()
print('x**6')
print()

print('y1 = {:.3f} + {:.3f}j'.format(y1.real, y1.imag))
print('y2 = {:.3f} + {:.3f}j'.format(y2.real, y2.imag))
print('y3 = {:.3f} + {:.3f}j'.format(y3.real, y3.imag))
print('y4 = {:.3f} + {:.3f}j'.format(y4.real, y4.imag))
print('y5 = {:.3f} + {:.3f}j'.format(y5.real, y5.imag))

print()
print('(x-1)**6')
print()

print('z1 = {:.3f} + {:.3f}j'.format(z1.real, z1.imag))
print('z2 = {:.3f} + {:.3f}j'.format(z2.real, z2.imag))
print('z3 = {:.3f} + {:.3f}j'.format(z3.real, z3.imag))
print('z4 = {:.3f} + {:.3f}j'.format(z4.real, z4.imag))
print('z5 = {:.3f} + {:.3f}j'.format(z5.real, z5.imag))

plt.plot([x1.real, x2.real, x3.real, x4.real, x5.real], [x1.imag, x2.imag, x3.imag, x4.imag, x5.imag], 'ro')


plt.plot([0, 1], [0, np.sqrt(3)], 'k--')
plt.plot([0, 1], [0, -np.sqrt(3)], 'k--')
plt.plot([0, 1], [0, np.sqrt(3)/3], 'k--')
plt.plot([0, 1], [0, -np.sqrt(3)/3], 'k--')

plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid()
plt.xlim([0, 1])
plt.ylim([-1, 1])
plt.show()
```
