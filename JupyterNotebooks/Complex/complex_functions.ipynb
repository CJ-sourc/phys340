# Complex Functions Examples

One of the most important complex functions is the complex exponential function. The complex exponential function is defined as

$$ e^z = e^{x+iy} = e^x e^{iy} = e^x (\cos(y) + i \sin(y)) $$

where $z = x + iy$ is a complex number. The complex exponential function is implemented in the `numpy` module as `exp`. The function `exp` can be used to evaluate the complex exponential function at a complex number $z$ as follows:

```python
import numpy as np
z = 1 + 1j
w = np.exp(z)
```


````python
def cfunc(z):
    return z**2
    #return np.exp(z)
````

````python
def plot_complex(z, f):
    
    xmin = 1.2*min(f.real)-0.5
    xmax = 1.2*max(f.real)+0.5
    ymin = 1.2*min(f.imag)-0.5
    ymax = 1.2*max(f.imag)+0.5
    
    xm = 1.2*min(z.real)-0.5
    xM = 1.2*max(z.real)+0.5
    ym = 1.2*min(z.imag)-0.5
    yM = 1.2*max(z.imag)+0.5

    fig, ax = plt.subplots(1, 2, figsize=(10,5))
    ax[0].plot(z.real,z.imag,'b.', label = 'z')
    ax[1].plot(f.real,f.imag,'g.', label = 'f(z)')
    
    # plot axes
    ax[0].plot([xm,xM],[0,0],'k')
    ax[0].plot([0,0],[ym,yM],'k')
    ax[1].plot([xmin,xmax],[0,0],'k')
    ax[1].plot([0,0],[ymin,ymax],'k')
    
    # Set limits
    ax[0].set_xlim([xm,xM])
    ax[0].set_ylim([ym,yM])
    ax[1].set_xlim([xmin,xmax])
    ax[1].set_ylim([ymin,ymax])
    
    # Set labels
    ax[0].set_xlabel('Real(z)')
    ax[0].set_ylabel('Imaginary(z)')
    ax[1].set_xlabel('Real(f(z))')
    ax[1].set_ylabel('Imaginary(f(z))')
    
````

````python
x = np.linspace(0.0,2,100)
y = x
z = x + 1j*y

f = cfunc(z)

plot_complex(z,f)
````

````python
theta = np.linspace(0,np.pi,40)
x = np.cos(theta)
y = np.sin(theta)
z = x + 1j*y

f = cfunc(z)
plot_complex(z,f)
````

# Complex Gamma Function

The complex Gamma function is defined as

$$  \Gamma(z) = \int_0^\infty t^{z-1} e^{-t} dt $$

where $z$ is a complex number. The complex Gamma function is implemented in the `scipy.special` module as `gamma`. The function `gamma` can be used to evaluate the complex Gamma function at a complex number $z$ as follows:

```python
from scipy.special import gamma
z = 1 + 1j
G = gamma(z)
```


````python
# Complex Gamma Function
from scipy.special import gamma, factorial


x = np.linspace(-4.5,4.5,50)
y = np.linspace(-4.5,4.5,50)
X,Y = np.meshgrid(x,y)
Z = X + 1j*Y

# Evaluate the complex Gamma function
G = gamma(Z)

absG = np.abs(G)

# Make a 3D plot of the function
#fig = plt.figure(figsize=(10,5))
#ax = fig.add_subplot(1, 2, 1, projection='3d')
#ax.scatter(X, Y, absG, c = absG, cmap = 'jet')


# Use plotly to make a 3D plot of the function
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=absG)], 
                layout=go.Layout(title='Complex Gamma Function'), 
                layout_scene = dict(
                    xaxis = dict(nticks=10, range=[-4.5,4.5],),
                    yaxis = dict(nticks=10, range=[-4.5,4.5],),
                    zaxis = dict(nticks=6, range=[0,6],)),
                layout_showlegend=True, 
                layout_width=1000, 
                layout_height=700
                )
fig.show()
````
