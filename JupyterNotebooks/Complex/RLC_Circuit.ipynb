```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
v0, omega, R, L, C, t = sym.symbols('v0 omega R L C t', real=True, positive=True)

V = v0*sym.exp(sym.I*omega*t)
ZR = R
ZL = sym.I*omega*L
ZC = 1/(sym.I*omega*C)

Z = ZR + ZL + ZC

I = V/Z

I = sym.simplify(I)
I = sym.expand(I)

sym.pprint(I)
```

```python
t = np.linspace(0, 0.02, 1000)
v0 = 10
omega = 1000.0
#omega = 7071.0678
R = 100
L = 2e-3
C = 10e-6

V = v0*np.exp(1j*omega*t)
ZR = R
ZL = 1j*omega*L
ZC = 1/(1j*omega*C)

Z = ZR + ZL + ZC

print('Complex impedance = ', Z)

I = V/Z

plt.plot(t, np.real(I), label='real')
plt.plot(t, np.imag(I), label='imag')
plt.xlabel('t')
plt.ylabel('I')
plt.legend()
plt.show()
```

```python
plt.plot(t, np.imag(V), label='V(t)')
plt.plot(t, 100*np.imag(I), label='100*I(t)')
plt.xlabel('t')
plt.ylabel('I/V')
plt.legend()
plt.show()
```

```python
omega_range = np.linspace(100, 100000, 100)
Q = []
for omega_val in omega_range:
    V = v0*np.exp(1j*omega_val*t)
    Z = R + 1j*omega_val*L + 1/(1j*omega_val*C)
    I = V/Z
    Q.append(R*abs(I)/abs(V))
```

```python
plt.plot(omega_range, Q)
plt.xlabel('omega')
plt.ylabel('Q')
plt.show()
```
