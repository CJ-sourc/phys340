```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate 400 points on a circle of radius 1
# and center (0,0)

npts = 400

theta = np.linspace(0, 2*np.pi, npts)
x = np.cos(theta)
y = np.sin(theta)

# plot the points
plt.scatter(x, y, s=0.5)
```

```python
# Consider the following transformation matrix

A = np.array([[0, 1], [-2, -3]])

# Apply the transformation to the points
# by matrix multiplication

x2 = []
y2 = []
for i in range(npts):
    v = np.array([[x[i]], [y[i]]])
    v2 = A @ v
    x2.append(v2[0, 0])
    y2.append(v2[1, 0])
```

```python
# plot the original and the transformed points
point_choice1 = 129
point_choice2 = 349

fig = plt.figure(figsize=(8, 8))
plt.scatter(x, y, label='original', s=0.5)
plt.scatter(x2, y2, label='transformed', s=0.5)

# draw a line from the origin to the point
plt.plot([0, x[point_choice1]], [0, y[point_choice1]], 'b')
plt.plot([0, x2[point_choice1]], [0, y2[point_choice1]], 'b--')

plt.plot([0, x[point_choice2]], [0, y[point_choice2]], 'r')
plt.plot([0, x2[point_choice2]], [0, y2[point_choice2]], 'r--')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.legend()


print('original point 1: ({:.2f}, {:.2f})'.format(x[point_choice1], y[point_choice1]))
print('original point 2: ({:.2f}, {:.2f})'.format(x[point_choice2], y[point_choice2]))
```

```python
xa = np.linspace(-4,4,npts)
ya = 0.90/-0.44*xa
yb = -0.71/0.71*xa

# calculate transformed points
x2a = []
y2a = []
for i in range(npts):
    v = np.array([[xa[i]], [ya[i]]])
    v2 = A @ v
    x2a.append(v2[0, 0])
    y2a.append(v2[1, 0])
    
x2b = []
y2b = []
for i in range(npts):
    v = np.array([[xa[i]], [yb[i]]])
    v2 = A @ v
    x2b.append(v2[0, 0])
    y2b.append(v2[1, 0])

plt.scatter(xa,ya,s=0.5)
plt.scatter(x2a,y2a, s=0.5)
plt.scatter(xa,yb, s=0.5)
plt.scatter(x2b,y2b, s=0.5)
```
