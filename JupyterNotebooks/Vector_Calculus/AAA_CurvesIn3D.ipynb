# Curves in 3D

Curves, or "paths" in 3D are defined by a parametric equation of the form:

$$x = x(t)$$
$$y = y(t)$$
$$z = z(t)$$

where $t$ is a parameter that varies between $t_{min}$ and $t_{max}$.

The parametric equation can be written in vector form as:

$$\vec{r}(t) = x(t)\hat{i} + y(t)\hat{j} + z(t)\hat{k}$$

The vector $\vec{r}(t)$ is called the position vector of the curve.

The derivative of the position vector is called the tangent vector:

$$\vec{v}(t) = \frac{d\vec{r}}{dt} = \frac{dx}{dt}\hat{i} + \frac{dy}{dt}\hat{j} + \frac{dz}{dt}\hat{k}$$

## Example 1

Plot the curve defined by the parametric equations:

$$x = 2t$$
$$y = 3t^2$$
$$z = 4t^3$$

for $t$ between $-1$ and $1$.

   


```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib notebook

t = np.linspace(-1,1,30)
x = 2*t
y = 3*t**2
z = 4*t**3

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x,y,z,color='b',lw=3,label='Curve')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Curve in 3D')

# draw coordinate axes
ax.plot([-2,2],[0,0],[0,0],color='k',linestyle='--')
ax.plot([0,0],[-2,2],[0,0],color='k',linestyle='--')
ax.plot([0,0],[0,0],[-2,2],color='k',linestyle='--')

# changed the viewing angle
ax.view_init(30,30)

plt.show()
```

## Tangent Vector

The tangent vector to the curve is given by the derivative of the position vector:

$$\vec{v}(t) = \vec{\nabla} \cdot \vec{r}(t)$$

where $\nabla$ is the so-called "del" , or gradient operator:

$$\nabla = \hat{i} \frac{d}{dt} + \hat{j} \frac{d}{dt} + \hat{k} \frac{d}{dt}$$

Thus, we have that:

$$\vec{v}(t) = \frac{d\vec{r}}{dt} = \frac{dx}{dt}\hat{i} + \frac{dy}{dt}\hat{j} + \frac{dz}{dt}\hat{k}$$

## Example 2

Plot the curve defined by the parametric equations:

$$x = 2\cos(\pi t)$$
$$y = 3\sin(\pi t)$$
$$z = t$$

for $t$ between $0$ and $10$.  Also, plot the tangent vector to the curve, at points along the curve (i.e. at different values of $t$).  

The tangent vector is given by:

$$\vec{v}(t) = \frac{d\vec{r}}{dt} = \frac{dx}{dt}\hat{i} + \frac{dy}{dt}\hat{j} + \frac{dz}{dt}\hat{k}$$

where:

$$\frac{dx}{dt} = -2\pi\sin(\pi t)$$
$$\frac{dy}{dt} = 3\pi\cos(\pi t)$$
$$\frac{dz}{dt} = 1$$

```python
t = np.linspace(0,10,100)
x = 2*np.cos(np.pi*t)
y = 3*np.sin(np.pi*t)
z = t

# Calulate the tangent vector
dxdt = -2*np.pi*np.sin(np.pi*t)
dydt = 3*np.pi*np.cos(np.pi*t)
dzdt = np.ones_like(t)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x,y,z,color='b',lw=3,label='Curve')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Curve in 3D')

# draw coordinate axes
ax.plot([-2,2],[0,0],[0,0],color='k',linestyle='--')
ax.plot([0,0],[-2,2],[0,0],color='k',linestyle='--')
ax.plot([0,0],[0,0],[-2,2],color='k',linestyle='--')

# Set the x, y, z limits
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_zlim(0,10)

# changed the viewing angle
ax.view_init(30,30)

# Plot the tangent vector
ax.quiver(x,y,z,dxdt,dydt,dzdt,color='g',label='Tangent vector',length=0.1)

plt.show()
```
