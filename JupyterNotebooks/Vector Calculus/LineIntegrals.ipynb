# Line Integrals of Vector Fields

In physics, we often want to calculate the work done by a force. For example, if we have a force $\vec{F}$ acting on a particle that moves along a curve $C$, then the work done by the force is given by the line integral

$$W = \int_C \vec{F} \cdot d\vec{r}$$

where $d\vec{r}$ is the differential displacement vector along the curve $C$. If we write $d\vec{r} = dx\,\vec{i} + dy\,\vec{j} + dz\,\vec{k}$, then we can write the line integral as

$$W = \int_C F_x\,dx + F_y\,dy + F_z\,dz$$

where $F_x$, $F_y$, and $F_z$ are the components of the force $\vec{F}$.

In this section, we will learn how to calculate line integrals of vector fields. We will also learn how to calculate the work done by a force along a curve.

## Line Integrals of Vector Fields

Let $\vec{F}(x,y,z) = P(x,y,z)\,\vec{i} + Q(x,y,z)\,\vec{j} + R(x,y,z)\,\vec{k}$ be a vector field defined on a curve $C$ parametrized by $x = x(t)$, $y = y(t)$, and $z = z(t)$, where $a \leq t \leq b$. Then the line integral of $\vec{F}$ along $C$ is given by

$$\int_C \vec{F} \cdot d\vec{r} = \int_a^b \vec{F}(x(t),y(t),z(t)) \cdot \vec{r}'(t)\,dt$$

where $\vec{r}(t) = x(t)\,\vec{i} + y(t)\,\vec{j} + z(t)\,\vec{k}$ is the position vector of a point on $C$.

### Example

Evaluate the line integral $\int_C \vec{F} \cdot d\vec{r}$, where $\vec{F}(x,y,z) = x\,\vec{i} + y\,\vec{j} + z\,\vec{k}$ and $C$ is the curve given by the vector function $\vec{r}(t) = t\,\vec{i} + t^2\,\vec{j} + t^3\,\vec{k}$, where $0 \leq t \leq 1$.

### Solution

First, we calculate $\vec{F}(\vec{r}(t))$ and $\vec{r}'(t)$:

$$\vec{F}(\vec{r}(t)) = \vec{F}(t,t^2,t^3) = t\,\vec{i} + t^2\,\vec{j} + t^3\,\vec{k}$$

$$\vec{r}'(t) = \vec{i} + 2t\,\vec{j} + 3t^2\,\vec{k}$$

Then we calculate $\vec{F}(\vec{r}(t)) \cdot \vec{r}'(t)$:

$$\vec{F}(\vec{r}(t)) \cdot \vec{r}'(t) = (t\,\vec{i} + t^2\,\vec{j} + t^3\,\vec{k}) \cdot (\vec{i} + 2t\,\vec{j} + 3t^2\,\vec{k}) = t + 2t^3 + 3t^5$$

Finally, we calculate the line integral:

$$\int_C \vec{F} \cdot d\vec{r} = \int_0^1 \vec{F}(\vec{r}(t)) \cdot \vec{r}'(t)\,dt = \int_0^1 t + 2t^3 + 3t^5\,dt = \frac{1}{2} + \frac{1}{2} + \frac{3}{6} = \frac{3}{2}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0,1,20)
x = t
y = t**2
z = t**3

dxdt = np.ones_like(t)
dydt = 2*t
dzdt = 3*t**2

ax.plot(x,y,z, color='b', lw=4, label="C")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

F = np.array([x,y,z]).T
r = np.array([x,y,z]).T

ax.quiver(F[:,0], F[:,1], F[:,2], r[:,0], r[:,1], r[:,2], color='g', lw=1,label="F(r)")
# Plot the tangent vector
ax.quiver(x,y,z,dxdt,dydt,dzdt,color='r',label="r'",length=0.1)
ax.legend()

# calculate the dr vector tangent to the curve
dr = np.array([dxdt,dydt,dzdt]).T

# calculate the F.dr vector
Fdr = np.sum(F*dr, axis=1)

# Calculate the line integral
W = np.trapz(Fdr, t)

print('The line integral is W = {:.2f}'.format(W))


```
