# Fourier Series

The Fourier series is a way to represent a periodic function as a sum of sines and cosines. The Fourier series is given by:

$$f(t) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left( a_n \cos\left(\frac{2\pi n t}{T}\right) + b_n \sin\left(\frac{2\pi n t}{T}\right) \right)$$

where $T$ is the period of the function and $a_0$, $a_n$, and $b_n$ are the Fourier coefficients. The Fourier coefficients are given by:

$$a_0 = \frac{2}{T} \int_{0}^{T} f(t) dt$$

$$a_n = \frac{2}{T} \int_{0}^{T} f(t) \cos\left(\frac{2\pi n t}{T}\right) dt$$

$$b_n = \frac{2}{T} \int_{0}^{T} f(t) \sin\left(\frac{2\pi n t}{T}\right) dt$$

The Fourier series can be used to represent any periodic function, and the coefficients can be found using the integrals above.

# Alternative Form

In the form above, we wrote the integrals over a single period, from $0$ to $T$. However, we could also integrate over $two$ periods, from $-T$ to $T$. This would give us the following alternative form for the Fourier series:

$$f(t) = a_0 + \sum_{n=1}^{\infty} \left( a_n \cos\left(\frac{2\pi n t}{T}\right) + b_n \sin\left(\frac{2\pi n t}{T}\right) \right)$$

where the Fourier coefficients are now given by:

$$a_0 = \frac{1}{T} \int_{-T}^{T} f(t) dt$$

$$a_n = \frac{1}{T} \int_{-T}^{T} f(t) \cos\left(\frac{2\pi n t}{T}\right) dt$$

$$b_n = \frac{1}{T} \int_{-T}^{T} f(t) \sin\left(\frac{2\pi n t}{T}\right) dt$$

Why is this a userful form?  Well, it is often easier to calculate the integrals over a symmetric range, from $-T$ to $T$, than over a single period, from $0$ to $T$. This is because the sines and cosines are symmetric functions, and so the integrals over a symmetric range are often zero. Specifically, the sine function is an $odd$ function on the interval $-T$ to $T$, and the cosine function is an $even$ function on the interval $-T$ to $T$.  If the function $f(t)$ is also symmetric, then the integrals of the product of the sine or cosine with $f(t)$ may be zero, by inspection.

This can simplify the calculation of the Fourier coefficients.

# Determination of the Fourier Coefficients

Let's understand where the expressions for the Fourier coefficients come from.  We will start with the expression for $a_0$.

## Calculation of a_0

The average value of a function, $f(t)$, over a single period, from $0$ to $T$, is given by:

$$\frac{1}{T} \int_{0}^{T} f(t) dt$$

Now, the average value of any sine or cosine function over a single period is zero. This is because the sine and cosine functions are symmetric about the origin, and so the positive and negative parts of the function cancel each other out when averaged over a single period.  This is true for any $n$, and so we can write:

$$\frac{a_0}{2} = \frac{1}{T} \int_{0}^{T} f(t) dt$$

$$a_0 = \frac{2}{T} \int_{0}^{T} f(t) dt$$

## Calculation of a_n and b_n

Beginning with the expression for $f(t)$, we can multiply both sides by $\cos\left(\frac{2\pi m t}{T}\right)$ and integrate over a single period, from $0$ to $T$. This gives:

$$\int_{0}^{T} f(t) \cos\left(\frac{2\pi m t}{T}\right) dt = \frac{a_0}{2} \int_{0}^{T} \cos\left(\frac{2\pi m t}{T}\right) dt + \\
\sum_{n=1}^{\infty} \left( a_n \int_{0}^{T} \cos\left(\frac{2\pi m t}{T}\right) \cos\left(\frac{2\pi n t}{T}\right) dt + \\
b_n \int_{0}^{T} \sin\left(\frac{2\pi m t}{T}\right) \cos\left(\frac{2\pi n t}{T}\right) dt \right)$$

Now, the integral next to the $a_0$ term is zero.

Looking at the integral next to the $b_n$ we see that the product of an odd function and an even function is an odd function, and so the integral is zero.

Looking at the integral next to the $a_n$ term, we can use the identity:

$$\cos(a)\cos(b) = \frac{1}{2} \left( \cos(a-b) + \cos(a+b) \right)$$

to write the integral as:

$$\int_{0}^{T} \cos\left(\frac{2\pi m t}{T}\right) \cos\left(\frac{2\pi n t}{T}\right) dt = \frac{1}{2} \int_{0}^{T} \cos\left(\frac{2\pi (m-n) t}{T}\right) dt + \frac{1}{2} \int_{0}^{T} \cos\left(\frac{2\pi (m+n) t}{T}\right) dt$$

The first integral is zero if $m \neq n$, and the second integral is zero if $m \neq -n$.  So, the only non-zero term is when $m=n$, and the integral is:

$$\int_{0}^{T} \cos\left(\frac{2\pi m t}{T}\right) \cos\left(\frac{2\pi m t}{T}\right) dt = \frac{T}{2}$$

So, finally, we have:

$$a_m = \frac{2}{T} \int_{0}^{T} f(t) \cos\left(\frac{2\pi m t}{T}\right) dt$$


# Example 1: Sine Wave

The Fourier series can be used to represent a sine wave, of frequency $f$. The equation of this sine wave is:

$$f(t) = \sin(2\pi f t) = \sin(\frac{2\pi t}{T})$$

With this in mind, we would expect to find that $a_0 = 0$, $a_n = 0$, and $b_n = 1$ for $n=1$, and zero for all $n>1$.

The period is $T = 1/f$, so the Fourier coefficients are:

$$a_0 = \frac{2}{T} \int_{0}^{T} \sin(\frac{2\pi t}{T}) dt$$
$$ = \frac{2}{T} \frac{T}{2\pi} \left[ -\cos(\frac{2\pi t}{T} \right]_0^T$$
$$ = \frac{1}{\pi} \left[ -1 - (-1) \right] = 0$$

Now, we could have made this easier by using the alternative form of the Fourier series, and integrating from $-T$ to $T$.  In this case, the integral would have been zero by inspection, because the sine function is an odd function!

$$a_n = \frac{2}{T} \int_{0}^{T} \sin(\frac{2\pi t}{T}) \cos\left(\frac{2\pi n t}{T}\right) dt$$

Here is another situation where it is helpful to use the alternative form of the Fourier series.  Because we have an integral of an odd function times an even function,from $-T$ to $T$, the integral will be zero, for any $n$!

$$b_n = \frac{2}{T} \int_{0}^{T} \sin(\frac{2\pi t}{T}) \sin\left(\frac{2\pi n t}{T}\right) dt$$

Now we have the product of two even functions, and so we cannot solve this by inspection, unfortunately.  To solve it analytically, we could use double angle formulas to expand the section integrand function and go from there.  Or, we can evaluate this integral using Wolfram Alpha, and the result is:

$$b_n = \delta_{n1}$$

## Example 2: Square Wave

The Fourier series can also be used to represent a square wave. The equation of this square wave is:

$$f(t) = \begin{cases} 1 & \text{if } 0 \leq t < T/2 \\ -1 & \text{if } T/2 \leq t < T \end{cases}$$

Note that this is an $odd$ function, and so the $a_0$ and $a_n$ coefficients will be zero.

The period is $T$, so the Fourier coefficients are:

$$a_0 = \frac{2}{T} \int_{0}^{T} f(t) dt = 0$$

$$a_n = \frac{2}{T} \int_{0}^{T} f(t) \cos\left(\frac{2\pi n t}{T}\right) dt = 0$$

$$b_n = \frac{2}{T} \int_{0}^{T} f(t) \sin\left(\frac{2\pi n t}{T}\right) dt$$

$$ = \frac{2}{T} \int_{0}^{T/2} \sin\left(\frac{2\pi n t}{T}\right) dt - \frac{2}{T} \int_{T/2}^{T} \sin\left(\frac{2\pi n t}{T}\right) dt $$

$$= \frac{2}{T} \left( \frac{-T}{2\pi n} \cos\left(\frac{2\pi n t}{T}\right) \right) \Bigg|_{0}^{T/2} - \frac{2}{T} \left( \frac{-T}{2\pi n} \cos\left(\frac{2\pi n t}{T}\right) \right) \Bigg|_{T/2}^{T} = $$

$$= \frac{2}{T} \left( \frac{-T}{2\pi n} \cos\left(\frac{2\pi n T}{2T}\right) - \frac{-T}{2\pi n} \cos\left(0\right) \right) - \frac{2}{T} \left( \frac{-T}{2\pi n} \cos\left(\pi n\right) - \frac{-T}{2\pi n} \cos\left(\frac{2\pi n T}{2T}\right) \right) = $$



$$=\frac{4}{n\pi} \text{ for } n \text{ odd}$$

So, the Fourier series for a square wave is:

$$f(t) = \sum_{n=1,3,5,...}^{\infty} \frac{4}{n\pi} \sin\left(\frac{2\pi n t}{T}\right)$$

$$ = \frac{4}{\pi} \sin(\frac{2\pi t}{T}) + \frac{4}{3\pi} \sin(\frac{6\pi t}{T}) + \frac{4}{5\pi} \sin(\frac{10\pi t}{T}) + ...$$

```python
# How could we calculate the Fourier coefficients for a sine wave (Example 1) or a square wave (Example 2), using sympy?

import sympy as sym

# Define the symbols
t = sym.symbols('t')
n = sym.symbols('n', integer=True)
T = sym.symbols('T', real=True, positive=True)
pi = sym.pi

T = T.subs(T, 1)

# Define the functions
sine_wave = sym.sin(2*pi*t/T)
square_wave = sym.Piecewise((1, t < T/2), (-1, t >= T/2))
    
choice = 2

if choice == 1:
    function = sine_wave
else:
    function = square_wave
    
# Calculate the Fourier coefficients
a0 = 2/T*sym.integrate(function, (t, 0, T))

an = 2/T*sym.integrate(function*sym.cos(2*n*pi*t/T), (t, 0, T))

bn = 2/T*sym.integrate(function*sym.sin(2*n*pi*t/T), (t, 0, T))

print(f'The Fourier coefficients are:')
print(f'a0:')
sym.pprint(a0)
print(f'an:')
sym.pprint(an)
print(f'bn:')
sym.pprint(bn)

bn_array = [bn.subs(n, i) for i in range(1, 13)]
print(f'bn_array:')
sym.pprint(bn_array)
```

```python
# Construct and plot the function

import numpy as np
import matplotlib.pyplot as plt

# Define the functions

def sine_wave(t, T):
    return np.sin(2*np.pi*t/T)

def square_wave(t, T):
    if t % T < T/2:
        return 1
    else:
        return -1
    
# Create an array of time values
t = np.linspace(-0.1, 3.1, 1000)

# Create an array of square wave values
if choice == 1:
    f = np.array([sine_wave(ti, 1) for ti in t])
else:
    f = np.array([square_wave(ti, 1) for ti in t])

# Plot the square wave
plt.plot(t, f)
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
plt.title('Function')
plt.show()
```

```python
# Calculate the Fourier series for the square wave

# Define the Fourier series function
def fourier_series(t, T, n_max):
    global a0, an, bn
    an_fourier = [an.subs(n, i) for i in range(1,n_max+1)]
    bn_fourier = [bn.subs(n, i) for i in range(1,n_max+1)]
    
    f = a0/2
    for ni in range(1, n_max+1):
        f +=  an_fourier[ni-1] * np.cos(2*np.pi*ni*t/T) + bn_fourier[ni-1] * np.sin(2*np.pi*ni*t/T) 
    
    return f

# choose period
T = 1

# Choose n
n_max = 10

# Create an array of Fourier series values
f_fourier_3 = np.array([fourier_series(ti, T, n_max) for ti in t])

# Plot the Fourier series and the square wave
plt.plot(t, f)
plt.plot(t, f_fourier_3)
plt.xlabel('Time (s)')
plt.ylabel('f(t)')
plt.title(f'Fourier Series for Square Wave (n={n_max})')
plt.show()
```

```python
# Plot the Fourier Components for the square wave

# Create an array of Fourier series values
xa = np.linspace(0, n_max, n_max+1)
xb = np.linspace(1, n_max, n_max)

ana = [an.subs(n, i) for i in range(0,n_max+1)]
bna = [bn.subs(n, i) for i in range(1,n_max+1)]

# Plot the Fourier components
plt.stem(xa, ana, 'r.',label='a_n')
plt.stem(xb, bna, 'b.',label='b_n')
plt.xlabel('n')
plt.ylabel('Fourier Coefficient')
plt.title('Fourier Components for Square Wave')
plt.legend()
    
```

If we know the exact form of the function, $f(t)$, we can calculate the Fourier coefficients, and use them to calculate the Fourier series. This can be used to represent the function as a sum of sines and cosines.

Again, we can simply evaluate the Fourier coefficients to as large a value of $n$ as we like, and the Fourier series will converge to the original function. This is a powerful result, and is the basis for many applications of the Fourier series in engineering and physics.

However, it is very important to note that as we increase the value of $n$, we are also increasing the frequency of the sines and cosines in the Fourier series. In the real world, we are often limited by the frequency response of our measurement devices, and so we may not be able to measure the high frequency components of a signal. This is a very important consideration when using the Fourier series in practice.
