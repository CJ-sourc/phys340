# Fourier Series

## Sine and Cosine Series

The Fourier series of a function $f(x)$ is given by (for $-L \leq x \leq L$), i.e. the alternative form of the Fourier series shown in AAA_FourierSeries.ipynb:

$$  f(x) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos \left( \frac{n \pi x}{L} \right) + b_n \sin \left( \frac{n \pi x}{L} \right) \right] $$

where

$$ a_0 = \frac{1}{2L} \int_{-L}^{L} f(x) dx $$

$$ a_n = \frac{1}{L} \int_{-L}^{L} f(x) \cos \left( \frac{n \pi x}{L} \right) dx $$

$$ b_n = \frac{1}{L} \int_{-L}^{L} f(x) \sin \left( \frac{n \pi x}{L} \right) dx $$

and $L$ is the period of the function.


## Example: Square Wave

Let's use a Fourier series to approximate a square wave with a period of L.

The square wave is given by

$$ f(x) = \begin{cases} 1 & \text{if } 0 \leq x < 0.5L \\ -1 & \text{if } 0.5L \leq x < L \end{cases} $$

The real Fourier series of the square wave, as demonstrated in AAA_FourierSeries.ipynb, is given by

$$f(x) = \sum_{n=1,3,5,...}^{\infty} \frac{4}{n\pi} \sin\left(\frac{2\pi n x}{L}\right)$$

where $L$ is the period of the function.

Let's now use the complex Fourier series to approximate the square wave.

$$c_n = \frac{1}{2L} \int_{-L}^{L} f(x) e^{-i n \pi x/L} dx $$

where $c_n$ are the complex Fourier coefficients.

As before, we can break this into four integrals:

$$c_n = \frac{1}{2L} \int_{-L}^{-0.5L} e^{-i n \pi x/L} dx + \frac{1}{2L} \int_{-0.5L}^{0} -e^{-i n \pi x/L} dx + \frac{1}{2L} \int_{0}^{0.5L} e^{-i n \pi x/L} dx + \frac{1}{2L} \int_{0.5L}^{L} -e^{-i n \pi x/L} dx $$

Evaluating these integrals, we find that

$$c_n = \frac{1}{2L} \left[ \frac{e^{-i n \pi x/L}}{-i n \pi/L} \right]_{-L}^{-0.5L} - \frac{1}{2L} \left[ \frac{e^{-i n \pi x/L}}{-i n \pi/L} \right]_{-0.5L}^{0} + \frac{1}{2L} \left[ \frac{e^{-i n \pi x/L}}{-i n \pi/L} \right]_{0}^{0.5L} - \frac{1}{2L} \left[ \frac{e^{-i n \pi x/L}}{-i n \pi/L} \right]_{0.5L}^{L} $$

$$c_n = \frac{i}{2n\pi} \left (  \left[ e^{-i n \pi x/L} \right]_{-L}^{-0.5L} - \left[ e^{-i n \pi x/L} \right]_{-0.5L}^{0} + \left[ e^{-i n \pi x/L} \right]_{0}^{0.5L} - \left[ e^{-i n \pi x/L} \right]_{0.5L}^{L} \right )$$

$$c_n = \frac{i}{2n\pi} \left (  e^{-i n \pi (-0.5L)/L} - e^{-i n \pi (-L)/L} - e^{-i n \pi (0)/L} + e^{-i n \pi (-0.5L)/L} + e^{-i n \pi (0.5L)/L} - e^{-i n \pi (0)/L} - e^{-i n \pi (L)/L} + e^{-i n \pi (0.5L)/L} \right )$$

$$c_n = \frac{i}{2n\pi} \left (  e^{i n \pi/2} - e^{i n \pi} - e^{0} + e^{i n \pi/2} + e^{-i n \pi/2} - e^{0} - e^{=i n \pi} + e^{-i n \pi/2} \right )$$

$$c_n = \frac{i}{n\pi} \left ( e^{i n \pi/2} + e^{-i n \pi - e^0 - e^{i n \pi}} \right )$$

Now, we can use that:

$$e^{i n \pi} = cos(n \pi) + i sin(n \pi) = (-1)^n$$
$$e^{-i n \pi} = cos(n \pi) - i sin(n \pi) = (-1)^n$$

$$e^{i n \pi/2} + e^{-i n \pi/2} = 2 cos(n \pi/2)$$

Thus, we find that

$$c_n = \frac{i}{n\pi} \left ( 2 cos(n \pi/2) - 1 - (-1)^n \right ) $$

From this, we can calculate the coefficients $c_n$ for $n = 1, 2, 3, ...$.

Let's calculate the coefficients $c_n$ for $n = 1, 2, 3, ...$ and compare them to the coefficients $a_n$ from the sine series.

$$c_1 = \frac{i}{\pi} \left ( 2 cos(\pi/2) - 1 - (-1) \right ) = \frac{i}{\pi} \left ( 0 - 1 + 1 \right ) = 0$$
$$c_2 = \frac{i}{2\pi} \left ( 2 cos(2\pi/2) - 1 - (-1)^2 \right ) = \frac{i}{2\pi} \left ( - 2 - 1 - 1 \right ) = -\frac{4i}{2\pi}$$
$$c_3 = \frac{i}{3\pi} \left ( 2 cos(3\pi/2) - 1 - (-1)^3 \right ) = \frac{i}{3\pi} \left ( 0 - 1 - (-1) \right ) = 0$$
$$c_4 = \frac{i}{4\pi} \left ( 2 cos(4\pi/2) - 1 - (-1)^4 \right ) = \frac{i}{4\pi} \left ( 2 - 1 - 1 \right ) = 0$$
$$c_5 = \frac{i}{5\pi} \left ( 2 cos(5\pi/2) - 1 - (-1)^5 \right ) = \frac{i}{5\pi} \left ( 0 - 1 - (-1) \right ) = 0$$
$$c_6 = \frac{i}{6\pi} \left ( 2 cos(6\pi/2) - 1 - (-1)^6 \right ) = \frac{i}{6\pi} \left ( -2 - 1 - 1 \right ) = -\frac{4i}{6\pi} $$

Thus, we see that the coefficients $c_n$ are zero for $n$ odd, and are $-4i/(2n\pi)$ for $n$ even. 

We can see that the non-zero coefficients $c_n$ are the same (modulo the factor of $i$) as the coefficients $b_n$ from the sine series, and the coefficients $a_n$ from the cosine series are zero.

Thus, we see that the complex Fourier series is a more elegant way to represent the Fourier series, as it combines the sine and cosine series into a single series.









```python
import sympy as sym

x = sym.symbols('x')
n = sym.symbols('n', integer=True)
L = sym.symbols('L', real=True, positive=True)
pi = sym.pi

fpos = sym.Piecewise((1, x < L/2), (-1, x >= L/2))
fneg = sym.Piecewise((1, x < -L/2), (-1, x >= -L/2))

c_n = 1/(2*L) * sym.integrate(fpos * sym.exp(-1j * n * pi * x/L), (x, 0, L)) + 1/(2*L) * sym.integrate(fneg * sym.exp(-1j * n * pi * x/L), (x, -L, 0))

c_n = c_n.simplify()

print('c_n = ')
sym.pprint(c_n)

for i in range(-11, 11):
    print('c_{:d} = '.format(i), c_n.subs(n, i).evalf())
```

# Complex Fourier Series

Another way to write the Fourier series is in terms of complex exponentials.  Yes, this is the THIRD way that we have looked at solving the same problem.  And, yes, it is the most elegant way to solve the problem!

The complex Fourier series approximation of a function $f(x)$ is given by the $real$ part of

$$  f_{Fourier}(x) = \sum_{n=-\infty}^{\infty} c_n e^{i n \pi x/L} $$

where

$$ c_n = \frac{1}{2L} \int_{-L}^{L} f(x) e^{-i n \pi x/L} dx $$

and $L$ is the period of the function.

The coefficients $c_n$ are called the (complex) Fourier coefficients.  Notice that we only have a $SINGLE$ series, and not separate sine and cosine series, and we only have a $SINGLE$ set of coefficients!

Let's expand the regular Fourier series to see that it is equivalent to the complex Fourier series.

We use that  $\cos(\theta) = \frac{1}{2} \left( e^{i \theta} + e^{-i \theta} \right)$ and $\sin(\theta) = \frac{1}{2i} \left( e^{i \theta} - e^{-i \theta} \right)$.

Thus, we can write the Fourier series as

$$ f(x) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \frac{1}{2} \left( e^{i n \pi x/L} + e^{-i n \pi x/L} \right) + b_n \frac{1}{2i} \left( e^{i n \pi x/L} - e^{-i n \pi x/L} \right) \right] $$
$$ f(x) = a_0 + \sum_{n=1}^{\infty} \left[ \frac{a_n - i b_n}{2} e^{i n \pi x/L} + \frac{a_n + i b_n}{2} e^{-i n \pi x/L} \right] $$

Comparing this to the complex Fourier series, we see that the coefficients are related by

$$ c_0 = a_0 $$
$$ c_n = \frac{a_n - i b_n}{2} $$
$$ c_{-n} = \frac{a_n + i b_n}{2} $$
$$ a_n = c_n + c_{-n} $$
$$ b_n = i \left( c_n - c_{-n} \right) $$



```python
# Let's use a complex Fourier series to approximate a square wave with a period of 1 from 0.0 to 6.0, as in AAA_FourierSeries.ipynb

import numpy as np
import matplotlib.pyplot as plt

n_terms = 40

from scipy import signal

x = np.linspace(-0.0, 6.0, 1000)
period = 1
y = signal.square(2*np.pi * x/period)
        
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

# plot axes
plt.plot([-0.1, 6.1], [0, 0], 'k')
plt.plot([0, 0], [-1, 1], 'k')

def cn(n):
    global y, x, period
    c = y* np.exp(-1j * n * np.pi * x/period)
    return c.sum() / len(x) 

def f(x, Nh):
    global y, period
    f = np.array([cn(i) * np.exp(1j * i * np.pi * x/period) for i in range(-Nh-1, Nh+1)])
    return f.sum()

yf = np.array([f(t, n_terms) for t in x])

plt.plot(x, yf.real, 'r') 
```

```python
# Make a plot of the Fourier series in the complex plane

plt.plot(yf.real, yf.imag, 'b')
```

```python
# Make a plot of the Fourier coefficients (real and imaginary parts)

n = np.arange(1, n_terms+1)
c = np.array([cn(i) for i in n])

fig, ax = plt.subplots(1,2, figsize=(10, 5))

ax[0].plot(n, c.real, 'b')
ax[1].plot(n, c.imag, 'r')
ax[0].set_xlabel('n')
ax[0].set_ylabel('Real(c_n)')
ax[1].set_xlabel('n')
ax[1].set_ylabel('Imag(c_n)')
ax[0].grid()
ax[1].grid()
ax[0].set_ylim(-0.7, 0.7)
ax[1].set_ylim(-0.7, 0.7)
```

# Let's generate some time series data and see if we can use a Fourier series to approximate it

I want to emphasize, in the strongest possible terms, that we are specifying the physical frequencies here.  When we calculate the complex Fourier coefficients, we are calculating the coefficients of the complex exponentials, which by our definition are $e^{i n \pi x/L}$, i.e. they do NOT include the factor of 2. We have to remember this!

```python
f1 = 20.1
f2 = 45.3
f3 = 92.8

npts = 1001

x = np.linspace(0, 1, npts)
# Create a signal that is the sum of three sine waves of the given frequencies, all with amplitude 1
y = np.sin(2 * np.pi * f1 * x) + np.sin(2 * np.pi * f2 * x) + np.sin(2 * np.pi * f3 * x)

plt.plot(x, y)

print('Sampling frequency = ', 1/(x[1]-x[0]))
print('Nyquist frequency = ', 0.5/(x[1]-x[0]))
```

# Nyquist Frequency

The Nyquist frequency is the highest frequency that can be represented in a sampled signal.  It is given by

$$ f_{Nyquist} = \frac{1}{2 \Delta t} $$

where $\Delta t$ is the sampling interval.

The Nyquist frequency is half the sampling frequency, and is an important consideration in deciding how to construct the Fourier series.

If we have a time series with a sampling frequency of 1000 Hz, then the Nyquist frequency is 500 Hz.  Thus, we can only represent frequencies up to 500 Hz in the Fourier series.

Next, we have to think about the number of terms in the Fourier series.  The highest frequency that can be represented is given by the Nyquist frequency. Let this be represented by $n_{max}$. 

Thus, the highest frequency in the Fourier series is given by

$$ f_{max} = \frac{n_{max}}{L} $$

where $L$ is the period of the function.

Thus, the number of terms in the Fourier series is given by

$$ n_{terms} = n_{max} = f_{max} L $$

```python
# Let's use a complex Fourier series to approximate the time series data

f_nyquist = 500.0 # Nyquist frequency as calculated above
period = 1.0 # i.e. the total time interval of the data
n_terms = int(f_nyquist * period)

print('max frequency = ', n_terms/period)
print('min frequency = ', 1/period)

yf = np.array([f(t, n_terms) for t in x])

plt.plot(x, y)
plt.plot(x, 0.5*yf.real, 'r')
plt.xlim(0, 0.2)
```

```python
# Plot the Fourier coefficients

n = np.arange(1, n_terms+1)
freq = n/period/2 # N.B. the factor of 2 here!!!!!!!!!!!!!!
c = np.array([cn(i) for i in n])

fig, ax = plt.subplots(1,2, figsize=(10, 5))

ax[0].plot(freq, abs(c.real), 'b')
ax[1].plot(freq, abs(c.imag), 'r')
ax[0].set_xlabel('frequency')
ax[0].set_ylabel('Real(c_n)')
ax[1].set_xlabel('frequency')
ax[1].set_ylabel('Imag(c_n)')
ax[0].set_ylim(0, 0.6)
ax[1].set_ylim(0, 0.6)
ax[0].grid()
ax[1].grid()
ax[0].set_xlim(0, 150)
ax[1].set_xlim(0, 150)
```

# Average Power

We can calculate the power spectrum of a signal using the Fourier coefficients. The average power is given by

$$ P = \frac{1}{2L} \int_{-L}^{L} |f(x)|^2 dx = \sum_{n=-\infty}^{\infty} |c_n|^2 $$

where $L$ is the period of the function.  Thus, a plot of the power spectrum vs. frequency is given by plotting $|c_n|^2$ vs. $n/L$.

```python
# Let's plot the power spectrum

fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.plot(freq, c.real**2+c.imag**2, 'b')

ax.set_xlabel('frequency')
ax.set_ylabel('Power')
ax.set_xlim(0, 100)
```

```python
# Compute the FFT of the time series data

FT = np.fft.fft(y)
freqfft = np.fft.fftfreq(len(x), x[1]-x[0])

# Let's compare the power spectrum from the Fourier series with the power spectrum from the FFT

fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.plot(freqfft, abs(FT)**2/npts**2, 'r.')
ax.plot(freq, (c.real**2+c.imag**2), 'b')
ax.set_xlim(0, 100)
ax.set_xlabel('frequency (Hz)')
ax.set_ylabel('Power (normalized)')
```
